```python
8. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints
Greetings! if anything else is stored in spam.
```

```python
def ankit(spam):
    if spam == 1:
        print("Hello")
    elif spam == 2:
        print("Howdy")
    else:
        print("Greetings!")
    return

ankit(1)
ankit(2)
ankit(3)
ankit(0)
```

```python
 11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
    
ans:- These three statement will give same result.
    range(10), The upper bound is given 10 which is excluded. The default lower bound is 0 and default jump is 1.

In range(0,10), The lower bound is given 0 which is included and upper bound is given 10 which is excluded. 
Default jump is 1.

In range(0,10,1), The lower bound is 0 and the upper bound is 10 and the jump is 1.
```
